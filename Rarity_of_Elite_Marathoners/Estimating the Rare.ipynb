```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math as mt

# read in the google raw word data file
raw_words_file = open("google_vocab_rawwordfreqs.txt", mode = "r")

raw_words = []

for line in raw_words_file:
    ele = line.strip("\n")
    raw_words.append(float(ele))

# read in the corpus word data

vocab_file = open("vocab_cs_mod.txt", mode = "r")

corp_rank = []
count = []

logCR = []
logC = []

for line in vocab_file:
    eles = line.split(" ")
    scr = float(eles[0])
    corp_rank.append(scr)
    sc = float(eles[1].strip("\n"))
    count.append(sc)

    logCR.append(scr)
    logC.append(sc)

plt.loglog(logCR, logC, ".")
plt.show()
```

```python
import numpy as np

def N_k(k):
    import numpy as np
    return 3.46 * (10**8)*k**(-0.661)

more_freqs = N_k(np.arange(0,200))

new_freqs = (more_freqs[1:-2] - more_freqs[2:-1])
#print(np.arange(0, len(new_freqs)))
plt.loglog(np.arange(0, len(new_freqs)), new_freqs, '.')
plt.loglog(corp_rank, count, '.')
plt.xlabel('Rank')
plt.ylabel('Word Count')
plt.show()
```

```python
new_allFreqs = new_freqs.tolist()+count
all_ones = new_allFreqs[new_allFreqs == 1]
new_allFreqs = np.array(new_allFreqs)
```

```python
# proportion of words that only appear once
all_ones = new_allFreqs[new_allFreqs == 1]
np.sum(all_ones)/np.sum(len(new_allFreqs))
```

```python
new_allFreqs = new_freqs.tolist()+count
new_mean = sum(new_allFreqs)/sum(count)
print('mean', new_mean)
new_var = np.var(new_allFreqs)
print('variance', new_var)
```

```python
# number of unique words
np.sum(count)/np.sum(new_allFreqs)
```

```python
# the total number and fraction of words left out of Google's data set
print(np.sum(new_freqs.tolist()))
print(np.sum(new_freqs.tolist())/np.sum(new_allFreqs))
```
